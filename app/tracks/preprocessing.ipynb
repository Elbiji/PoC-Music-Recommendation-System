## Pembacaan Data

```python
import pandas as pd
import numpy as np

df = pd.read_csv('dataset.csv')
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def calculate_TF_IDF_df(df):
    df['combined_text'] = df['artists'].fillna(' ') + ' ' + df['album_name'].fillna(' ') + ' ' + df['track_name'].fillna(' ') + ' ' + df['track_genre'].fillna(' ')

    tfidf_vectorize = TfidfVectorizer(stop_words="english", max_features=1000)
    tfidf_matrix = tfidf_vectorize.fit_transform(df['combined_text'])

    feature_names = tfidf_vectorize.get_feature_names_out()
    tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

    df_with_tfidf = pd.concat([df, tfidf_df], axis=1)
    return df_with_tfidf

def calculate_TF_IDF_np(df):
    df['combined_text'] = df['artists'].fillna(' ') + ' ' + df['album_name'].fillna(' ') + ' ' + df['track_name'].fillna(' ') + ' ' + df['track_genre'].fillna(' ')

    tfidf_vectorize = TfidfVectorizer(stop_words="english", max_features=1000)
    tfidf_matrix = tfidf_vectorize.fit_transform(df['combined_text'])

    tfidf_df = tfidf_matrix.toarray()

    np.save("track_vectorize.npy", tfidf_df)

df_with_tfidf = calculate_TF_IDF_df(df)
    
```
